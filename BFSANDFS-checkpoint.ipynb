### Task 2
Implement the Breadth First Search (BFS) both iterative and recursive algorithm

```python
MyGraph={
    'A' : ['B','C'],
    'B' : ['A','D','E'],
    'C' : ['A','F'],
    'D' : ['B'],
    'E' : ['B','F'],
    'F' : ['C','E']
}
def BFS_Iterative_Alogorith(MyGraph,root):
    Queue, visited = [root],[]
    while Queue:
        vertex = Queue.pop(0)
        for N in MyGraph[vertex]:
            if N not in visited:
                visited.append(N)
                Queue.append(N)
    return visited
print(BFS_Iterative_Alogorith(MyGraph,'A'))
```
