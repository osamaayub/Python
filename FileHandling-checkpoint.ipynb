# Representation of Search Problem Tasks

##### 8 Puzzle Problem:
Here, we have a 3Ã—3 matrix with movable tiles numbered from 1 to 8 with a blank space. The tile adjacent to the blank space can slide into that space. The objective is to reach a specified goal state similar to the goal state, as shown in the below figure.

In the figure, our task is to convert the current state into goal state by sliding digits into the blank space.

<b>The problem formulation:</b>

```python
IntialState:start with the matrix with the numbers 1 to 8 and a blank space.
Actions:We can only move blankspace up,down,left and right.
state space:we can apply dfs or bfs to solve the problem.
path:we can only make the blank_space move in four directions left,right,up and down but not diagonal.
GoalState/test:move the blank space to its correct place by sliding digits until the given goal state is reached.
solution:we will move the blank space left,right or up and down in different steps to reach the goal state.
TotalCost:cost to reach the goal state is 15 moves
pathcost: 1 per move

```

<b>8-queens problem:</b><br>
The aim of this problem is to place eight queens on a chessboard in an order where no queen may attack another. A queen can attack other queens either diagonally or in same row and column. From the following figure, we can understand the problem as well as its correct solution.

It is noticed from the above figure that each queen is set into the chessboard in a position where no other queen is placed diagonally, in same row or column. Therefore, it is one right approach to the 8-queens problem.

<b>The problem formulation:</b>

```python
IntialState: start with 0 queens on the chessboard
Actions:start placing the queen in the square of the chessboard
stateSpace:we need to place 8 queens on the chessboard not in same row,column and diagonal
path: we need to place only one queen in one row ,column and diagonal so they dont attack each other
GoalState/test:place 8 queens on chessboard sucessfully without each getting attacked
solution:We need to place 8 queens without attacking each other in the chessboard
TotalCost:11 per steps to place all the queens on  the chessboard
pathcost: 1 per move
```
