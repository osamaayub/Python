### Question 1:
Find the winning value of the Maximizer (Node A).


```python
#Implement Minimax Algorithm only for the winning value.
import math
def MinMaxAlgo(currentDepth,NodesIndex,Turns,Score,TargetDepth):
    
     if currentDepth==TargetDepth:
        return Score[NodesIndex]
     if(Turns):
        Max=max(MinMaxAlgo(currentDepth + 1,NodesIndex*2,False,Score,TargetDepth),MinMaxAlgo(currentDepth+1,NodesIndex*2+1,False,Score,TargetDepth))
        return Max
     else:
        Min=min(MinMaxAlgo(currentDepth+1,NodesIndex*2,True,Score,TargetDepth),MinMaxAlgo(currentDepth+1,NodesIndex*2+1,True,Score,TargetDepth))
        return Min
Terminal_values=[-1,4,2,6,-3,-5,0,7]
TreeDepth=int(math.log(len(Terminal_values),2))
print("Tree Depth is:",TreeDepth)
print("Winning Value of the Maximizier Node is:",end=" ")
print(MinMaxAlgo(0,0,True,Terminal_values,TreeDepth))
```

### Question 2:
Find the winning value of the maximizer along with the path.


```python
#Implement Minimax Algorithm that print paths to wining/optimal value.
import math
def MinMaxAlgo(currentDepth,NodesIndex,Turns,Score,TargetDepth):
    
     if currentDepth==TargetDepth:
        return Score[NodesIndex]
     if(Turns):
        return max(MinMaxAlgo(currentDepth + 1,NodesIndex*2,False,Score,TargetDepth),MinMaxAlgo(currentDepth+1,NodesIndex*2+1,False,Score,TargetDepth))
     else:
        return min(MinMaxAlgo(currentDepth+1,NodesIndex*2,True,Score,TargetDepth),MinMaxAlgo(currentDepth+1,NodesIndex*2+1,True,Score,TargetDepth))
Terminal_values=[-1,3,5,1,-6,-4,0,9]
TreeDepth=int(math.log(len(Terminal_values),2))
print("Tree Depth is:",TreeDepth)
print("Winning Value of the Maximizier Node is:",end=" ")
path=MinMaxAlgo(0,0,True,Terminal_values,TreeDepth)
print(path)
```
